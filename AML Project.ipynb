```python
def impute_basic(df, column):
    if df[column].skew() < 0.5:  # Relatively symmetric
        df[column].fillna(df[column].mean(), inplace=True)
    else:  # Contains outliers or skewed
        df[column].fillna(df[column].median(), inplace=True)

# Impute each column with missing values
for col in data_train.columns:
    if data_train[col].isnull().sum() > 0:
        impute_basic(data_train, col)
```
